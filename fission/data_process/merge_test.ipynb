```python
import pandas as pd

# 读取第一个CSV文件
abs_data = pd.read_csv('abs_regional_population_lga_2001_2021-2770189027992760124.csv')

# 读取第二个CSV文件
air_data = pd.read_csv('Final_air.csv')


```

```python

def preprocess_abs(abs_data):
    selected_rows = abs_data[abs_data[' state_name_2021'] == 'Victoria']
    
    selected_values = ['Wangaratta', 'Melton', 'Swan Hill', 'Mildura', 'Melbourne', 'Greater Bendigo', 'Greater Geelong', 'Whitehorse']
    filtered_df = selected_rows[selected_rows[' lga_name_2021'].isin(selected_values)]

    # Delte the non related column
    cols_to_drop = ['state_code_2021', ' lga_code_2021', ' state_name_2021', ' primaryindex']
    filtered_df = filtered_df.drop(cols_to_drop, axis=1)

    # Change the word 
    filtered_df.iloc[0, 2] = 'Bendigo'
    filtered_df.iloc[1, 2] = 'Geelong South'
    filtered_df.iloc[2, 2] = 'Melbourne CBD'
    filtered_df.iloc[7, 2] = 'Box Hill'

    return filtered_df
```

```python
abs_df = preprocess_abs(abs_data)
abs_df
```

```python
air_data 
```

```python
merged_df = air_data.merge(abs_df, how='inner', left_on='location_name', right_on=' lga_name_2021')
merged_df
```

```python
merged_df.to_csv('merged.csv', index=False)
```
